### Сезонность и тренды

Функция для вывода тренда сезонности и остатка

```python
def check_seasonal(data_in, name_col, period_in=0):
    if period_in == 0:
        decomposed = seasonal_decompose(data_in[name_col].dropna())
    else :
        decomposed = seasonal_decompose(data_in[name_col].dropna(), period=period_in)

    plt.figure(figsize=(8, 12))
    
    plt.subplot(311)
    decomposed.trend.plot(ax=plt.gca())
    plt.title('Trend')
    plt.xlabel('Дата')
    plt.ylabel('Превышающее количество заказов')
    
    plt.subplot(312)
    decomposed.seasonal.plot(ax=plt.gca()) 
    plt.title('Seasonality')
    plt.xlabel('Дата')
    plt.ylabel('Отклонение для сезонного количества заказов')
    
    plt.subplot(313)
    decomposed.resid.plot(ax=plt.gca()) 
    plt.title('Residuals')
    plt.tight_layout()
    plt.xlabel('Дата')
    plt.ylabel('Отклонение количество заказов')
    
    return decomposed
```

Добавление следующих признаков:
- день;
- неделя;
- год;
- день недели;
- значения предыдущих часов;
- скользящее среднее.

1) Задержка в сутки - 24ч;<br>
2) Скользящее среднее как сутки*2

```python
max_lag = hours_in_day
rolling_mean_size = hours_in_day*2
```

```python
def make_features(data_in, max_lag, rolling_mean_size):
    data = data_in.drop(columns=['mean', 'std'])
    
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(window=rolling_mean_size).mean() 
    return data
```

```python
data_work = make_features(data_hour, max_lag, rolling_mean_size)
```

## Обучение

Подготовка данных для обучения

```python
target_col = 'num_orders'
```

Разделение на тренировочную и тестовую выборки

```python
train, test = train_test_split(data_work, shuffle=False, test_size=0.1)
train = train.dropna()
```

```python
y_train = train[target_col]
X_train = train.drop(columns=[target_col])

y_test = test[target_col]
X_test = test.drop(columns=[target_col])
```

Поскольку все столбцы числовые, можем отобрать их все и провести масштабирование

```python
num_col_names = list(data_work.drop(columns=target_col, axis=1).columns)
```

```python
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train[num_col_names])
X_test_st = scaler.transform(X_test[num_col_names])
    
df_X_train_st = pd.DataFrame(X_train_st, columns=num_col_names)
df_X_test_st = pd.DataFrame(X_test_st, columns=num_col_names)
```

```python
X_train_fin = df_X_train_st
X_test_fin = df_X_test_st
```

Для сравнения будем рассматривать линейную регрессию и решающее дерево;<br>
Первой моделью будет линейная регрессия

### Линейная регрессия

```python
model_lr = LinearRegression() 

model_lr.fit(np.array(X_train_fin), np.array(y_train))
```

```python
tscv = TimeSeriesSplit(n_splits=5)
```

```python
RMSE_lin = -np.array(
    cross_val_score(
        model_lr,
        X_train_fin,
        y_train,
        cv=tscv,
        scoring='neg_root_mean_squared_error'
    )
).mean()

print("RMSE обучающей выборки:", RMSE_lin)
```

Далее рассмотрим решающее дерево

### Дерево

Константы для модели

```python
RANDOM_STATE = 42
num_col_names = list(X_train_fin.columns)
```

Пайплайн для решающего дерева

```python
pipe_final = Pipeline([
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

```python
param_grid = [
    # словарь для модели DecisionTreeRegressor()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 25),
        'models__max_features': range(2, 100),
        'models__min_samples_split': range(2, 50),
        'models__min_samples_leaf': range(2, 10)  
    }
]
```

Общий пайплайн

```python
%%time

randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=tscv,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1,
    random_state=RANDOM_STATE
)
```

Поиск лучшей модели

```python
randomized_search.fit(X_train_fin, y_train)
```

Параметры и метрика решающего дерева

```python
best_tree_linear = randomized_search.best_estimator_
```

```python
best_RMSE_tree = -randomized_search.best_score_
print("RMSE обучающей выборки:", best_RMSE_tree)
```
